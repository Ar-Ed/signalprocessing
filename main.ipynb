```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import cv2 as cv
```

#### Opening and Display Images

```python
# read image
lenna = cv.imread("Lenna.png")

# convert to grayscale image
gray_lenna = cv.cvtColor(lenna, cv.COLOR_BGR2GRAY)

# BGR 3 channel, gray_scale is 1 channel 
print(lenna.shape, gray_lenna.shape)
```

```python
#display images by using the opencv api 
cv.imshow("Lenna Original", lenna)
cv.imshow("Lenna Gray", gray_lenna)
cv.waitKey();
```

```python
# display using Matplotlib
plt.subplot(121)
plt.title("Original")
# converted image to RGB from BGR then displayed (try to see what happens when "plt.imshow(lenna)" is called)
plt.imshow(cv.cvtColor(lenna, cv.COLOR_BGR2RGB))

plt.subplot(122)
plt.title("Gray")
plt.imshow(gray_lenna, cmap="gray")
```
