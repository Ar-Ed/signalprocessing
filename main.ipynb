#### Opening and Display Images

```python
# read image
lenna = cv.imread("Lenna.png")

# convert to grayscale image
gray_lenna = cv.cvtColor(lenna, cv.COLOR_BGR2GRAY)

# BGR 3 channel, gray_scale is 1 channel 
print(lenna.shape, gray_lenna.shape)
```

```python
# display using Matplotlib
plt.subplot(121)
plt.title("Original")
# converted image to RGB from BGR then displayed (try to see what happens when "plt.imshow(lenna)" is called)
plt.imshow(cv.cvtColor(lenna, cv.COLOR_BGR2RGB))

plt.subplot(122)
plt.title("Gray")
plt.imshow(gray_lenna, cmap="gray")
```

#### Blurring Images
There are many methods to blur images (Gaussian, median, bilateral kernels etc.) but most of them use 2D Convolution (Kernels differ). Learn about 2D Convolution

For Further Reading about Blurring:
https://docs.opencv.org/4.x/d4/d13/tutorial_py_filtering.html

```python
# let's blur our lenna
kernel_size = (11,11)
lenna_blurred = cv.GaussianBlur(lenna, kernel_size, 0)

# BGR --> RGB
lenna_blurred_rgb = cv.cvtColor(lenna_blurred, cv.COLOR_BGR2RGB) 

#display gaussian blurred image
plt.imshow(lenna_blurred_rgb)
```

#### Applying Your Own Filters

You can define your own kernels or find different kernels on the internet and see what happens 

```python
# let's define a kernel or average filtering
kernel = np.ones((7,7)) / 49
display(kernel)

lenna_average_filtered = cv.filter2D(lenna, -1, kernel)

# BGR --> RGB
lenna_blurred_filtered_rgb = cv.cvtColor(lenna_average_filtered, cv.COLOR_BGR2RGB) 

# display average filterred image
plt.imshow(lenna_blurred_filtered_rgb)
```
